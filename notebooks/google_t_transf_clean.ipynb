```python
# Import pandas create a df from the csv
import pandas as pd

df = pd.read_csv("../data/raw/google_trends_migration.csv")

# Drop the isPartial column
df = df.drop(columns=["isPartial"])

df.head() 
```

```python
# Convert the date column to datetime and resample it into yearly with mean values
df['date'] = pd.to_datetime(df['date'])
df.set_index('date', inplace=True) # set date as index
df = df.resample("YE").mean()  

# rename 2010-01-01 to 2010
df.index = df.index.year

# round values 
df = df.round(3)

df.head()
```

```python
# pivot the df to be like eurostat
df_pivot = df.T
df_pivot.head()
```

```python
# save the df as csv
df.to_csv("../data/processed/google_trends_migration_yearly.csv")
```
