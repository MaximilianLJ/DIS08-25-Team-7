### H1 - Migration trend
How has net migration to Germany developed between 2010 and 2024?
- Data source: Eurostat
- Measure: Crude net migration rate (per 1.000 inhabitants)
- Purpose:
  - Establish empirical context
  - Identify major fluctations and structural breaks

```python
# Load libraries and DataFrame
import pandas as pd
import matplotlib.pyplot as plt

# Transpose DataFrame and adjust type
df = pd.read_csv("../data/processed/combined_migration_data.csv")
df.columns = df.columns.str.strip()
df_t = df.set_index("Jahr").T
df_t = df_t.reset_index().rename(columns={"index": "Jahr"})
df_t["Jahr"] = df_t["Jahr"].astype(int)

df = df_t
df
```

```python
# H1
plt.figure(figsize=(10, 5))

# create line plot
plt.plot(
    df["Jahr"],
    df["Netto-Migrationsrate"],
    marker="o",
)
# customize ploit
plt.title("Netto Migrationsrate in Deutschland (2010-2024)")
plt.xlabel("Jahr")
plt.ylabel("Netto-Migrationsrate (pro 1.000 Einwohner)")
plt.grid(True)
plt.tight_layout()

# save figure
plt.savefig("../data/figures/netto_migrationsrate_deutschland.png", dpi=300, bbox_inches="tight")
print("saved figure")

plt.show()
```

### H1 Interpretation
The Above figure "Netto Migrationsrate in Deutschland (2010-2024)" shows the crude migration rate in Germany from 2010 - 2024. Net migration increased steadliy between 2010 and 2014 and reached a clear peak in 2015. This strong increase represents a major structural break and is linked to the European refugee crisis.  

After 2015, the net migration rate declined and remained lower until 2019. In 2020, migration decreased further, likely due to travel restrictions during the COVID-19 pandemic. From 2021 onwards, migration increased again, with a noticeble rise in 2023.  

Overall, net migration to Germany, in the chosen time frame, shows strong fluctuations, which are closely related to major events

### H2 Public opinion trend
How has public concern about migration changed between 2010 and 2024
- Data source: Selected Standard Eurobarometer 
- Measure: Share of respondents [%] who identify immigration as one of the two most important problems facing Germany
- Purpose: 
  - Describe long-term changes in public concern about migration
  - Identify periods of heightened or declining salience of migration in public opinion
  - Provide context for analysing the relationship between migration level and public attitudes


```python
# H2 drop all values besides "Meinungsprozentsatz" and "Jahr"
df_h2 = df[["Jahr", "Meinungsprozentsatz"]].dropna()
df_h2
```

```python

plt.figure(figsize=(10, 5))

plt.plot(
    df_h2["Jahr"],
    df_h2["Meinungsprozentsatz"],
    marker = "o"
)

plt.title("Meinungsprozentsatz zur Migration in Deutschland (2010-2024)")
plt.xlabel("Jahr")
plt.ylabel("Meinungsprozentsatz (%)")
plt.grid(True)

plt.tight_layout()

plt.savefig("../data/figures/öffentliche_meinung.png", dpi=300, bbox_inches="tight")
print("saved figure")
plt.show()
```

### H2 Interpretation
Figure "Meinungsprozentsatz zur Migration in Deutschland (2010 - 2024)" how public concern about migration in Germany changed between 2010 and 2024. Public concern was relatively low at the beginning of the period but increased strongly between 2013 and 2015, where also the highest recorded value lies.  

After 2015, the share of respondents who identified migration as a major problem declined steadily until 2022. This indicates a decreasing salience of migration in public opinion during the late 2010s and early 2020s. In 2023, public concern rose again, followed by slight decline in 2024.

Overall we can see the same conclusion as before that the concern for it rises with the amount of migration happening.

### H3 Relationship between migration levels and public concern  
Is there an empirical relationship between migration levels and public concern about migration in Germany?  
- Data Sources:
  - Eurostat
  - Eurobarometer
- Measures: 
  - Crude net migration rate
  - Share of respondents [%] who identify immigration as one of the two most important probelems facing Germany
- Purpose:
  - Asses whether higher migration levels are associated with higher public concern about migration
  - Provide an empirical basis for evaluating the interaction between migration dynamics and public attitudes

```python
df_h3 = df[["Jahr", "Netto-Migrationsrate", "Meinungsprozentsatz"]].dropna()
df_h3
```

```python
plt.figure(figsize=(10, 5))

plt.scatter(
    df_h3["Netto-Migrationsrate"],
    df_h3["Meinungsprozentsatz"],
)

plt.title("Zusammenhang zwischen Netto-Migrationsrate und Meinungsprozentsatz in Deutschland (2010-2024)")
plt.xlabel("Netto-Migrationsrate (pro 1.000 Einwohner)")
plt.ylabel("Meinungsprozentsatz (%)")
plt.grid(True)
plt.tight_layout()

plt.savefig("../data/figures/zusammenhang_migration_meinung.png", dpi=300, bbox_inches="tight")

plt.show()
```

### H3 Interpretaiton
This figure "Zusammenhang zwischen Netto-Migrationsrate und Meinungsprozentsatz in Deutschland (2010 - 2024)" depicts the relationship between crude net migration and public concern about migration in Germany from 2010 to 2024. 

The scatter plot suggests a positive relationship between migration levels and public concern about migration. However the relationship is not perfectly linear and there are noticeable variations between individual years. This indicates that there are more contributors than just the migration itself.

In summary, the results provide evidence of an association between migration levels and public concern about migration, but they do not allow for causal conclusions. 

### H4 Online attention and public concern
How does online attention to migration relate to public concern about migration in Germany?  
- Data sources:
  - Google Trends
  - Eurobarometer
- Measure: 
  - Google Trends search interest index (0-100) for migration related search terms
  - Share of respondents [%] who identify immigration as one of the two most important probelems facing Germany
- Purpose:
  - Approximate public and media attention to migration over time
  - Examine whether periods of heightened online attention coincide with increased public concern
  - Provide additional context for interpreting the relationship between migration levels and public attitudes

```python
df_h4 = df[["Jahr", "Meinungsprozentsatz", "Suchinteresse Migration", "Suchinteresse Flüchtlinge"]].dropna()
df_h4
```

```python
plt.figure(figsize=(10, 5))

plt.plot(
    df_h4["Jahr"],
    df_h4["Meinungsprozentsatz"],
    marker="o",
    label="Meinungsprozentsatz"
)

plt.plot(
    df_h4["Jahr"],
    df_h4["Suchinteresse Migration"],
    marker="o",
    label="Migration"
)

plt.plot(
    df_h4["Jahr"],
    df_h4["Suchinteresse Flüchtlinge"],
    marker="o",
    label="Flüchtlinge"
)

plt.title("Öffentliches Interesse an Migration in Deutschland (2010-2024)")
plt.xlabel("Jahr")
plt.ylabel("Prozentsatz / Suchinteresse")
plt.legend()
plt.grid(True)
plt.tight_layout()

plt.savefig("../data/figures/öffentliches_interesse_migration.png", dpi=300, bbox_inches="tight")
print("saved figure")
plt.show()
```

```python
from scipy.stats import pearsonr, spearmanr

# Migration
pearson_migration, pearson_migration_p = pearsonr(df_h4["Meinungsprozentsatz"], df_h4["Suchinteresse Migration"])
spearman_migration, spearman_migration_p = spearmanr(df_h4["Meinungsprozentsatz"], df_h4["Suchinteresse Migration"])

# Flüchtlinge
pearson_refugees, pearson_refugees_p = pearsonr(df_h4["Meinungsprozentsatz"], df_h4["Suchinteresse Flüchtlinge"])
spearman_refugees, spearman_refugees_p = spearmanr(df_h4["Meinungsprozentsatz"], df_h4["Suchinteresse Flüchtlinge"])

# a = 0.05
print("Migration: ")
print(f"Pearson: {pearson_migration}, p-value: {pearson_migration_p}")
print(f"Spearman: {spearman_migration}, p-value: {spearman_migration_p}")

print("Flüchtlinge: ")
print(f"Pearson: {pearson_refugees}, p-value: {pearson_refugees_p}")
print(f"Spearman: {spearman_refugees}, p-value: {spearman_refugees_p}")
```

### H4 – Interpretation

“Öffentliches Interesse an Migration in Deutschland (2010–2024)” illustrates the
relationship between online attention to migration-related topics and public
concern about migration in Germany between 2010 and 2024. Online attention is
measured using Google Trends data for the term “Migration” and for the often more
sensationalised term “Flüchtlinge” (“Refugees”).

Both the Spearman and Pearson correlation results indicate a much stronger
relationship between public concern and the search term “Flüchtlinge” compared to
the more general term “Migration”, especially during periods of heightened media
attention.

These findings suggest that public concern about migration is more closely linked
to attention toward refugee-related topics, which are often associated with
crisis situations and emotionally charged public debates. However, the results
should be interpreted with caution, as Google Trends measures search interest
rather than attitudes and does not allow for causal conclusions.
