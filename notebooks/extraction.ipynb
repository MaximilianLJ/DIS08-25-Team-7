### Scrape Data from Eurostat


```python
import requests
import pandas as pd

# URL generated from Eurostat (Copy API link)
url = "https://ec.europa.eu/eurostat/api/dissemination/sdmx/3.0/data/dataflow/ESTAT/demo_gind/1.0/*.*.*?c[freq]=A&c[indic_de]=CNMIGRAT,CNMIGRATRT,JAN,NATGROW&c[geo]=DE&c[TIME_PERIOD]=2010,2011,2012,2013,2014,2015,2016,2017,2018,2019,2020,2021,2022,2023,2024&compress=false&format=csvdata&formatVersion=1.0&lang=en&labels=label_only"

response = requests.get(url)
response.raise_for_status() # Safety check

# Save as .csv in the ./data/raw folder
df = pd.read_csv(url) 
df.to_csv("../data/raw/eurostat_demo_gind_DE_2010_2024.csv", index=False)
```

### Scrape Data from Google Trends

```python
from pytrends.request import TrendReq
import pandas as pd

pytrends = TrendReq(hl = "de-DE", tz=360)

keywords = ["Asyl", "Flüchtlinge", "Migration", "Ausländer", "Remigration"]
pytrends.build_payload(keywords, cat = 0, timeframe = "2010-01-01 2024-12-31", geo = "DE")

df_trends = pytrends.interest_over_time()
df_trends.to_csv("../data/raw/google_trends_migration.csv")
```
